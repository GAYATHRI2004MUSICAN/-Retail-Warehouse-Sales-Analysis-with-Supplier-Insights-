```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Warehouse_and_Retail_Sales.csv")

df.head()
df.info()
df.describe()
```

```python
df.columns
```

```python
df['DATE'] = pd.to_datetime(df['YEAR'].astype(str) + '-' + df['MONTH'].astype(str))
```

```python
monthly_sales = df.groupby('DATE')['RETAIL SALES'].sum()

monthly_sales.plot()
plt.title("Monthly Retail Sales Trend")
plt.show()
```

Sales Trend Insight: 
“Retail sales show a steady upward trend over the months, indicating consistent business growth and increasing customer demand.”

```python
item_type_sales = df.groupby('ITEM TYPE')['RETAIL SALES'].sum()

item_type_sales.plot(kind='bar')
plt.title("Sales by Item Type")
plt.show()
```

Item Type Insight: “Item Type Y generates the highest revenue, indicating customer preference towards this category and potential for further expansion.”

```python
supplier_sales = df.groupby('SUPPLIER')['RETAIL SALES'].sum().sort_values(ascending=False).head(10)

supplier_sales.plot(kind='bar')
plt.title("Top Suppliers by Sales")
plt.show()
```

Top Supplier Insight: “Supplier X contributes the highest share of total sales, suggesting strong market presence and high product demand from this supplier.”

```python
df[['RETAIL SALES', 'WAREHOUSE SALES']].sum().plot(kind='bar')
plt.title("Retail vs Warehouse Sales")
plt.show()
```

Retail vs Warehouse Insight: “Retail sales significantly outperform warehouse sales, highlighting that direct customer purchases are the primary revenue driver.”

```python
transfer = df.groupby('DATE')['RETAIL TRANSFERS'].sum()

transfer.plot()
plt.title("Retail Transfers Trend")
plt.show()
```

Transfers Insight: “Retail transfers show fluctuations, suggesting uneven inventory movement and possible supply chain inefficiencies.”
